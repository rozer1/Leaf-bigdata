```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *

spark = SparkSession.builder.appName("projekt-leaf").getOrCreate()
            
schema = StructType([\
    StructField("Class", IntegerType(), True),\
    StructField("Specimen Number", IntegerType(), True),\
    StructField("Eccentricity", FloatType(), True),\
    StructField("Aspect Ratio", FloatType(), True),\
    StructField("Elongation", FloatType(), True),\
    StructField("Solidity", FloatType(), True),\
    StructField("Stochastic Convexity", FloatType(), True),\
    StructField("Isoperimetric Factor", FloatType(), True),\
    StructField("Maximal Indentation Depth", FloatType(), True),\
    StructField("Lobedness", FloatType(), True),\
    StructField("Average Intensity", FloatType(), True),\
    StructField("Average Contrast", FloatType(), True),\
    StructField("Smoothness", FloatType(), True),\
    StructField("Third moment", FloatType(), True),\
    StructField("Uniformity", FloatType(), True),\
    StructField("Entropy", FloatType(), True)
                    ])
df = spark.read.option("header", "true").schema(schema).csv("leaf.csv")

# ratings = spark.read.csv("leaf.csv", sep=",", header=False, )
# df.show(16)
# df.printSchema()
df = df.withColumnRenamed("Class", "label").drop("Specimen Number")
df.createOrReplaceTempView("df")
df.show(10)
df.count()
```

```python
df_train, df_eval = df.randomSplit([0.8, 0.2], 40)
```

```python
from pyspark.ml import feature

vect = feature.VectorAssembler(inputCols=df.columns[:], outputCol="feat")
df_train = vect.transform(df_train)
```

```python
df_train = df_train.select("label", "feat")
```

```python
scaler = feature.StandardScaler(inputCol="feat", outputCol="features")
scaler_t = scaler.fit(df_train)
df_train = scaler_t.transform(df_train)
```

```python
df_train.show()
```

```python
from pyspark.ml import classification

forest = classification.RandomForestClassifier(maxDepth=8, minInstancesPerNode=5, seed=42)
forest_t = forest.fit(df_train)
pred_train = forest_t.transform(df_train)
```

```python
from pyspark.ml import evaluation
evaluator = evaluation.MulticlassClassificationEvaluator(metricName="accuracy", labelCol="label", predictionCol="prediction")
evaluator.evaluate(pred_train)
```

```python
df_eval = vect.transform(df_eval)
df_eval = df_eval.select("label", "feat")
df_eval = scaler_t.transform(df_eval)
pred_eval = forest_t.transform(df_eval)
```

```python
accuracy = evaluator.evaluate(pred_eval)
print("Accuracy = %s" % (accuracy))
print("Test Error = %s" % (1.0 - accuracy))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import itertools

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=90)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```
